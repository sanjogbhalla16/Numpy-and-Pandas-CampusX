### Selecting rows from a DataFrame

1. iloc - searches using index position
2. loc - searches using index labels

```python
# single row

movies.iloc[0]
```

```python
movies.iloc[0:5]
```

```python
# Fancy Indexing
movies.iloc[[0,4,5]]
```

```python
# loc will work with labels

students.loc['vansh']
```

```python
students.loc['sanjog':'vansh':2]
```

```python
# Fancy Indexing

students.loc[['sanjog','akash','vansh']]
```

Filtering a DataFrame

```python
ipl.head(2)
```

```python
# Find all the final winners of all seasons

mask = ipl['MatchNumber'] == 'Final'
new_df = ipl[mask]
new_df[['Season','WinningTeam']]
```

```python
# single line
ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
# how many super over finishes have occurred

ipl[ipl['SuperOver'] == 'Y']
```

```python
# how many matches has csk won in kolkata

ipl[(ipl['WinningTeam'] == 'Chennai Super Kings') & (ipl['City'] == 'Kolkata')].shape[0]
```

```python
# toss winner is match winner in percentage

win = ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]

total = ipl.shape[0]

percent_win = (win/total)*100
percent_win
```
